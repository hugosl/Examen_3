```python
from pylab import *
%matplotlib inline
```

```python
def derivada(f,x): #funcion derivada numerica
    h = 1e-8 #Un número suficientemente pequeño 
    d = (f(x + h) - f(x)) / h  # derivada numerica
    return d 
```

```python
def funcion_prueba(x): # funcion de prueba
    return x**3 - 5*x + 2
def derivada_analitica(x): # derivada analitica de la funcion de prueba 
    return 3*x**2 -5
```

```python
x=linspace(-60,60) # intervalo 
subplots(figsize=(12,12))
title('funcion prueba = rojo, derivada analitica = amarillo, derivada numerica = azul ',fontsize=16)
sol = derivada(funcion_prueba,x)
plot(x,sol,'-',color='blue')
plot(x,funcion_prueba(x),'-',color='red')
plot(x,derivada_analitica(x),'.',color='yellow')
grid(True)
show()
```
