```python
from pylab import *
%matplotlib inline
from scipy.special import * # biblioteca de funciones especiales
```

```python
x=linspace(0,5*pi) #intervalo del dominio de la funcion de Bessel
subplots(figsize=(10,6))
xlim(0,5*pi)
for i in range(4): # orden de la funcion de Bessel (0,1,2,3)
    y = jv(i,x) 
    plot(x,y,'-') 
    grid(True)
print "J0(x) = azul"
print "J1(x) = verde"
print "J2(x) = rojo"
print "J3(x) = azul cielo"    
show()
```
